# Feature Engineering
1. What is a parameter?
- A parameter is a variable used by a machine learning model to make predictions. Parameters are learned automatically during training.
   - Example:
    In linear regression:
         y = mx + c
   - Here:
      - m and c are parameters



2. What is correlation? What does negative correlation mean?
- Correlation measures the relationship between two variables.
  - Positive correlation: both viariables increase together.
  - Negative correlation: one variable increases while the other decreases

4. How does loss value help in determining whether the model is good or not?
- Loss value measures how much error exists between predicted output and actual output.
    - Lower loss = better model
    - Higher loss = poor        predictions
- Loss helps improve the model during training.

5. What are continuous and categorical variables?
- Continuous Variables:
   Variables that can take any numeric value.
  - Examples:
    - Height
    - Weight
    - Salary
- Categorical Variables:
  Variables representing categories or labels.
  Examples:
     - Gender
     - Color
     - City
